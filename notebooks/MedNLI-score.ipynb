```python
import os
import json
import pandas as pd

from pathlib import Path
```

```python
NAME_DATA = 'mednli'

PATH_BASE = Path('../')
PATH_MODELS = PATH_BASE/'models'
```

```python
dirs = [_ for _ in os.listdir(PATH_MODELS) if '.' not in _]
models = tuple(set(''.join(_.split('_')[1:][:-1]) for _ in dirs if NAME_DATA in _))

models
```

```python
args = []

for NAME_MODEL in dirs:
    # if models[0] not in NAME_MODEL:
    #     continue
    data = {}
    with open(PATH_MODELS/NAME_MODEL/'args.json', 'r') as f:
        d = json.load(f)
        data.update(d)
    with open(PATH_MODELS/NAME_MODEL/'result.csv', 'r') as f:
        d = pd.read_csv(f).iloc[0].to_dict()
        data.update(d)
    data['run_name'] = NAME_MODEL
        
    args.append(data)
```

```python
argsDF = pd.DataFrame(args)
df = argsDF.loc[
    :, 
    [
        'run_name',
        'num_train_epochs',
        'per_device_train_batch_size',
        'per_device_eval_batch_size',
        'gradient_accumulation_steps',
        'eval_accumulation_steps',
        'learning_rate',
        'weight_decay',
        'warmup_ratio',
        'warmup_steps',
        'evaluation_strategy',
        'metric_for_best_model',
        'load_best_model_at_end',
        'eval_loss',
        'eval_accuracy',
        'eval_precision',
        'eval_recall',
        'eval_f1'
    ]
]

df
```

```python
df.to_csv(PATH_MODELS/'args.csv', index=False)
```
